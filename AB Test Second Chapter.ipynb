## 利用卡方检验，去判别因子之间的独立性
## 原假设H0：各个因子之间的误差具有随机性
## 备择假设H1：因子之间的误差是具有人为性质


```python
from sklearn.feature_selection import chi2

Fdata = data[['user_id', 'source', 'device', 'operative_system', 'test', 'city']]
Fdata = Fdata.set_index(['user_id'])

Fdata
```

```python
le = LabelEncoder()

for item in Fdata.columns:
    Fdata[item] = le.fit_transform(Fdata[item])
    
Fdata
```

```python
colnames = ['source', 'device', 'operative_system', 'city']

chi2values, pvalues = chi2(Fdata.drop(['test'], axis = 1), Fdata.test)
pd.DataFrame({'Chi2': chi2values, 'Pvalue': pvalues}, index = colnames)
```
